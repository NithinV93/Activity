```python
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('diabetes_model.csv')
```

```python
pd.DataFrame(data)
```

# Countplot-Outcome

```python
warnings.filterwarnings('ignore')
```

```python
sns.countplot(data['Outcome'])
```

# Boxplot-Effect of age on outcome

```python
sns.boxplot(x='Outcome',y='Age',data=data).set_title('Effect of age on outcome')
```

# Chances of diabetes, if common in family using pedigree function

```python
sns.swarmplot(x='Outcome', y='DiabetesPedigreeFunction',data=data)
plt.title('Chances of diabetes, if common in family using pedigree function ')
```

# Insulin glucose Relation

```python
plt.scatter(data['Insulin'],data['Glucose'])
plt.title('Insulin-Glucose')
plt.xlabel('Insulin')
plt.ylabel('Glucose')
```

# Heatmap

```python
plt.figure(figsize=(10,10))
sns.heatmap(data.corr(),annot=True)
```
