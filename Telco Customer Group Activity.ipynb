```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset=pd.read_csv('Telco-Customer-Churn.csv')
```

```python
dataset
```

# 1.Compare churn count with respect to gender

```python
plt.figure(figsize=(10,10))
sns.countplot(x='Churn',data=dataset,hue='gender')
plt.title('Churn count VS Gender')
plt.grid()
dataset['Churn'].value_counts()
```

# 2. Find out how many female senior citizens there in the dataset 

```python
fsc=dataset[(dataset['gender']=='Female') & (dataset['SeniorCitizen']==1)]
```

```python
fsc
```

```python
fsc['SeniorCitizen'].count()
```

# 3.Compare Tenure with Total Charges

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x='tenure',y='TotalCharges',data=dataset,hue='SeniorCitizen')
plt.title('Tenure VS Total Charges',fontsize=14)
plt.grid()
```

```python
#From Scatterplot we infer that tenure and total charges are directly propotional, i.e as tenure increases, total charges also goes up.
```

# 4.Find out which contract preferred by the senior citizen. 

```python
sc=dataset[(dataset['SeniorCitizen']==1)]
```

```python
sc
```

```python
con_pre=sc.groupby(['SeniorCitizen','Contract']).sum()
```

```python
con_pre
```

```python
plt.figure(figsize=(8,8))
sns.countplot(x='Contract',data=sc)
```

```python
#It is inferred that Senior Citizens prefer Month-to-month contracts follwed by one year and two year contracts.
```

# 5.Comment on your findings on Payment Method

```python
plt.figure(figsize=(10,10))
sns.countplot(x='PaymentMethod',hue='SeniorCitizen',data=dataset)
plt.xticks(rotation=45)
plt.title('Payment Method', fontsize=14)
plt.grid()
```

```python
#Non-Senior and Senior Citizens prefer Electronic Check over all other payment methods.
#Senior Citizens prefer Mailed Check the least.
#Non-Senior Citizens prefer Credit card the least.
```
